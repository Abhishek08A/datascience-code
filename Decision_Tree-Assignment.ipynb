```python
df=df.fillna(df.mean())
```

```python
x=df.drop('num',axis=1)
y=df['num']
```

```python
x
```

```python
y
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
x_train,x_test,y_train,y_test
```

```python
x_train.shape,x_test.shape,y_train.shape,y_test.shape
```

```python
clf=DecisionTreeClassifier(random_state=42)
clf
```

```python
clf.fit(x_train,y_train)
```

```python
y_pred=clf.predict(x_test)
y_pred
```

```python
print('\nClassification Report:')
print(classification_report(y_test,y_pred))
```

```python
print('\nConfusion Matrix:')
print(confusion_matrix(y_test,y_pred))
```

```python
print("\nAccuracy:", accuracy_score(y_test, y_pred))
```

```python
param_grid = {
    'criterion': ['gini', 'entropy'],
    'max_depth': [None, 10, 20, 30],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 5]
}
```

```python
grid_search = GridSearchCV(DecisionTreeClassifier(random_state=42), param_grid, cv=5, scoring='accuracy')
grid_search.fit(x_train, y_train)
```

```python
print("\nBest Parameters from Grid Search:")
print(grid_search.best_params_)
```

```python
best_clf = grid_search.best_estimator_
best_clf.fit(x_train, y_train)
```

```python
y_pred_tuned = best_clf.predict(x_test)
y_pred_tuned
```

```python
print("\nClassification Report (Tuned Model):")
print(classification_report(y_test, y_pred_tuned))
```

```python
print("\nConfusion Matrix (Tuned Model):")
print(confusion_matrix(y_test, y_pred_tuned))
```

```python
print("\nAccuracy (Tuned Model):", accuracy_score(y_test, y_pred_tuned))
```

```python
plt.figure(figsize=(20, 10))
plot_tree(best_clf, feature_names=x.columns, class_names=[str(label) for label in y.unique()], filled=True, rounded=True)
plt.title("Decision Tree Structure")
plt.show()
```

```python
importances = pd.DataFrame({'Feature': x.columns, 'Importance': best_clf.feature_importances_})
importances = importances.sort_values(by='Importance', ascending=False)
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(x='Importance', y='Feature', data=importances)
plt.title("Feature Importance")
plt.show()
```

# Interview Questions:

```python
1. What are some common hyperparameters of decision tree models, and how do they affect the model's performance?
Ans: Hyperparameter Tunnings	           Description	                                 Effect on Performance
max_depth:-	 Maximum depth of the tree	-  Controls overfitting. Shallow = underfit, deep = overfit
min_samples_split:-	Minimum number of samples required to split a node	- Higher value → more conservative tree (less complex)
min_samples_leaf:-	Minimum number of samples required in a leaf node -	Prevents creation of small leaf nodes; reduces overfitting
max_leaf_nodes:-	Maximum number of leaf nodes	Limits complexity by bounding the total number of final decisions (leaves)
max_features:-	Number of features to consider when splitting a node	Introduces randomness → useful in ensembles (e.g., Random Forest)
```

```python
2. What is the difference between the Label encoding and One-hot encoding?
Ans:   Label Encoding
Definition:- Converts categories into numeric labels (e.g., "red" → 0, "blue" → 1)
Output Format:-	Single column with integer values
Use Case:-	For ordinal data (where order matters), or tree-based models (like Decision Trees, Random Forests)
Example:-	["cat", "dog", "mouse"] → [0, 1, 2]
Problem:-	Can mislead linear models into thinking one category is greater than another (e.g., "dog" > "cat")



2. One-Hot Encoding
Definition:-	Converts each category into a new binary column
Output Format;-	Multiple columns, each for a category
Use Case:-	For nominal data (no order), and models that are sensitive to numerical magnitude (e.g., Logistic Regression, SVM, Neural Nets)
Example:-	["cat", "dog", "mouse"] → [[1,0,0], [0,1,0], [0,0,1]]
Problem:-	Increases dimensionality — especially with high-cardinality categorical variables
```
