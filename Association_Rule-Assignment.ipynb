```python
data = {
    'TransactionID': [1, 2, 3, 4],
    'Products': [
        ['shrimp', 'almonds', 'avocado'],
        ['vegetables mix', 'green grapes', 'whole wheat flour'],
        ['yams', 'cottage cheese', 'energy drink'],
        ['tomato juice', 'low fat yogurt', 'green tea', 'honey']
    ]
}
```

```python
df = pd.DataFrame(data)
df
```

```python
all_products = ['shrimp', 'almonds', 'avocado', 'vegetables mix', 'green grapes', 'whole wheat flour', 
                'yams', 'cottage cheese', 'energy drink', 'tomato juice', 'low fat yogurt', 'green tea', 
                'honey', 'salad', 'mineral water', 'salmon', 'antioxidant juice', 'frozen smoothie', 
                'spinach', 'olive oil']
```

```python
def one_hot_encode(transaction_list, all_products):
    return [1 if product in transaction_list else 0 for product in all_products]
```

```python
basket_encoded = df['Products'].apply(lambda x: one_hot_encode(x, all_products))
```

```python
basket_encoded_df = pd.DataFrame(basket_encoded.tolist(), columns=all_products)
```

```python
basket_encoded_df['TransactionID'] = df['TransactionID']
```

```python
basket_encoded_df.set_index('TransactionID', inplace=True)
basket_encoded_df
```

```python
df.columns
```

```python
!pip install mlxtend
```

```python
from mlxtend.frequent_patterns import apriori, association_rules
frequent_itemsets = apriori( basket_encoded_df,min_support=0.05, use_colnames=True)
```

```python
rules = association_rules(frequent_itemsets, metric="lift", min_threshold=1,num_itemsets=5)
```

```python
filtered_rules = rules[(rules['confidence'] >= 0.7) & (rules['lift'] >= 1)]
```

```python
print("Filtered Association Rules:")
filtered_rules
```

```python
for index, row in filtered_rules.iterrows():
    print(f"Rule {index+1}: {row['antecedents']} -> {row['consequents']}")
    print(f"Support: {row['support']}, Confidence: {row['confidence']}, Lift: {row['lift']}")
    print("-" * 50)
```

# Interview Questions:

1.Lift is a metric used to evaluate the strength of an association rule. It measures how much more likely the consequent (right-hand side) of a rule is to occur when the antecedent (left-hand side) occurs, compared to when the antecedent and consequent are independent of each other.

#Lift helps to understand whether the rule is actually meaningful or if the antecedent and consequent occur together by chance.
#A higher lift value indicates a stronger association, which makes the rule more interesting and actionable for business decisions.


2.Support measures the frequency or proportion of transactions that contain both the antecedent and consequent of a rule.
#Confidence(A→B)=Support(A)/Support(A∩B)
​

#3. Computational Complexity:
Apriori algorithm, the most common algorithm for association rule mining, can be computationally expensive, especially for large datasets. This is because it needs to scan the entire dataset multiple times to count itemsets, which makes it slow for large numbers of transactions and items

#Association rules generally focus on binary relationships (e.g., “If A, then B”). More complex relationships, such as those involving multiple items or those requiring more sophisticated machine learning models, are not well-captured by traditional association rule mining.
