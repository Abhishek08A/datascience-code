```python
x=df1.drop(['class'],axis=1)
x
```

```python
y=df1['class']
y
```

```python
# Identify categorical and numerical columns
categorical_cols = x.select_dtypes(include=['object']).columns
numeric_cols = x.select_dtypes(include=['int64', 'float64']).columns
```

```python
# Encode only categorical columns
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
x_encoded = x.copy()
for col in categorical_cols:
    x_encoded[col] = le.fit_transform(x[col])  
x_encoded
```

```python
y_encoded = le.fit_transform(y)
y_encoded
#It’s a 1D array, and that’s perfectly valid for machine learning models like SVM.
#But if you're printing it using print(y_encoded), it might look like one long line.

```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x_encoded,y_encoded,train_size=0.8,random_state=42)
```

```python
from sklearn.svm import SVC
model = SVC()
model.fit(x_encoded, y_encoded)
```

```python
y_pred=model.predict(x_train)
y_pred
```

```python
from sklearn.metrics import classification_report
```

```python
from sklearn.metrics import accuracy_score
```

```python
print(accuracy_score(y_train,y_pred))
```

```python
print(classification_report(y_train,y_pred))
```

```python
# To check for the testing data set
y_pred1=model.predict(x_test)
y_pred1
```

```python
print(accuracy_score(y_test,y_pred1))
```

```python
print(classification_report(y_test,y_pred1))
```

```python
#Class 0: edible mushrooms
#Class 1: poisonous mushrooms

#Low Overall Accuracy (0.68)
#The model is biased toward class 1 because class 1 has more samples (267 vs 133).
#Weighted average F1-score is only 0.55, meaning the model's overall quality is weak despite high recall for class 1.

#Strong Performance on Class 1 Only
#For class 1:
#Recall = 1.00: All actual class 1 samples were correctly predicted.
#F1-score = 0.80: Good balance for class 1, but at the cost of failing on class 0.
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay

# Generate the confusion matrix
cm = confusion_matrix(y_test, y_pred1)

# Plot the heatmap
plt.figure(figsize=(6, 4))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=['Edible', 'Poisonous'], yticklabels=['Edible', 'Poisonous'])
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix - SVM Classification')
plt.show()
```

```python
 #Severe Misclassification of Edible Mushrooms

#Only 3 out of 133 edible mushrooms were correctly classified.
#130 edible mushrooms were incorrectly predicted as poisonous.

#Perfect Detection of Poisonous Mushrooms
#All 267 poisonous mushrooms were correctly identified.
#No poisonous mushrooms were misclassified as edible.
```

```python
#The SVM model is heavily biased toward predicting "poisonous".
#While this minimizes risk of poisoning, it also means many safe mushrooms are unnecessarily rejected.
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.svm import SVC
```

```python
# Define parameter grid
param_grid = {
    'kernel': ['rbf'],
    'C': [0.1, 1],
    'gamma': ['scale'],
    'class_weight': ['balanced']
}
```

```python
# Fit to training data
grid_search.fit(x_train, y_train)
```

```python
# Best parameters
print("🔍 Best Parameters:", grid_search.best_params_)
```

```python
# Evaluate on test data
y_pred2= grid_search.predict(x_test)
y_pred2
```

```python
print(accuracy_score(y_test,y_pred2))
```

```python
print(classification_report(y_test,y_pred2))
```

```python
from sklearn.svm import SVC
from sklearn.metrics import classification_report, accuracy_score
import pandas as pd

# Define kernels to test
kernels = ['linear', 'poly', 'rbf', 'sigmoid']
```

```python
#Store results
results = []
```

```python
for kernel in kernels:
    print(f"\n  Training SVM with kernel = '{kernel}'")
```

```python
model = SVC(kernel=kernel, class_weight='balanced')  # class_weight helps with imbalance
model.fit(x_train, y_train)
y_pred3 = model.predict(x_test)
y_pred3
```

```python
print(accuracy_score(y_test,y_pred3))
```

```python
print(classification_report(y_test, y_pred3))
```

```python
results.append({'Kernel': kernel, 'Accuracy': accuracy_score})
```

```python
# Create a summary table
results_df = pd.DataFrame(results)
print("\n Kernel Comparison Summary:\n", results_df)
```

```python
#Sigmoid Kernel
#Rarely performs well for classification.
#Sometimes used for similarity-based problems or to mimic neural networks.
#Likely to give the lowest performance here.
```

```python
#Strength	Explanation
# Handles High-Dimensional Data Well	SVM is great with many features (after encoding), especially when classes are well-separated in feature space.
# Nonlinear Decision Boundaries	With kernels like RBF, SVM can model complex class boundaries – useful for mushroom data.


#Weakness	Explanation
#Sensitive to Class Imbalance	As seen in your confusion matrix, SVM ignored the edible class without using class_weight='balanced'.
# Poor Interpretability	Hard to understand why a particular prediction was made, especially with RBF or poly kernels.
#Performance Can Drop on Poorly Scaled or Noisy Features	Outliers in cap_diameter, stalk_height, or poor encoding can mislead SVM.
```

```python
#-------------------------------------When SVM Is a Great Choice--------------------------------------------------------------------
#Bioinformatics / Medical Diagnosis:- 	SVM handles high-dimensional data like gene expression or disease classification (e.g., cancer detection). 
#Security / Fraud Detection :-	        Excellent at modeling rare class patterns, especially with good kernel tuning.
#Image Recognition:-	                With proper feature extraction (e.g., HOG), SVM can be competitive for handwritten digit or face recognition tasks.
#Text Classification / Spam Detection:- Performs well with sparse data (e.g., TF-IDF vectors) and clear boundaries between classes.
```
