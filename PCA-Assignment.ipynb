```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score, davies_bouldin_score
```

```python
for column in df.select_dtypes(include=np.number).columns:
    plt.figure(figsize=(12, 5))
    plt.subplot(1, 2, 1)
    sns.histplot(df[column], kde=True, bins=30)
    plt.title(f"Histogram of {column}")

    plt.subplot(1, 2, 2)
    sns.boxplot(x=df[column])
    plt.title(f"Boxplot of {column}")
    plt.show()
```

```python
correlation_matrix = df.corr()
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title("Correlation Matrix")
plt.show()
```

```python
scaler = StandardScaler()
scaled_data = scaler.fit_transform(df.select_dtypes(include=np.number))
```

```python
pca = PCA()
pca_data = pca.fit_transform(scaled_data)
```

```python
explained_variance = pca.explained_variance_ratio_
cumulative_variance = np.cumsum(explained_variance)
plt.figure(figsize=(10, 6))
plt.plot(range(1, len(explained_variance) + 1), cumulative_variance, marker='o', linestyle='--')
plt.title('Cumulative Explained Variance')
plt.xlabel('Number of Principal Components')
plt.ylabel('Cumulative Explained Variance')
plt.show()
```

```python
optimal_components = np.argmax(cumulative_variance >= 0.95) + 1
print(f"Optimal number of components: {optimal_components}")
```

```python
pca = PCA(n_components=optimal_components)
pca_transformed_data = pca.fit_transform(scaled_data)
```

```python
kmeans = KMeans(n_clusters=3, random_state=42)
original_clusters = kmeans.fit_predict(scaled_data)
```

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(x=scaled_data[:, 0], y=scaled_data[:, 1], hue=original_clusters, palette='viridis')
plt.title("Clustering on Original Data")
plt.show()
```

```python
silhouette_orig = silhouette_score(scaled_data, original_clusters)
davies_bouldin_orig = davies_bouldin_score(scaled_data, original_clusters)
print(f"Silhouette Score (Original Data): {silhouette_orig:.2f}")
print(f"Davies-Bouldin Index (Original Data): {davies_bouldin_orig:.2f}")
```

```python
pca_clusters = kmeans.fit_predict(pca_transformed_data)
```

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(x=pca_transformed_data[:, 0], y=pca_transformed_data[:, 1], hue=pca_clusters, palette='viridis')
plt.title("Clustering on PCA Data")
plt.show()
```

```python
silhouette_pca = silhouette_score(pca_transformed_data, pca_clusters)
davies_bouldin_pca = davies_bouldin_score(pca_transformed_data, pca_clusters)
print(f"Silhouette Score (PCA Data): {silhouette_pca:.2f}")
print(f"Davies-Bouldin Index (PCA Data): {davies_bouldin_pca:.2f}")
```

```python
#Task 5: Comparison and Analysis:
#1. Comparison of Clustering Results (Original vs PCA-transformed)

#Original Dataset Clustering:
  #Clustering is performed on high-dimensional data.
  #The algorithm captures more variance and detail but may also include noise or irrelevant features
  #Performance depends heavily on feature scale, correlation, and redundancy.

#PCA-Transformed Dataset Clustering:
   # Dimensionality is reduced, focusing on directions (principal components) that capture the most variance.
    # Noise and less-informative features are filtered out.
    #Clustering is more efficient and often visually easier to interpret in 2D/3D space.

#2. Similarities and Differences in Clustering Results
#Similarities:
#Core structure of the clusters may still be preserved if the variance captured by top principal components is high (e.g., 95%+).
#If data is well-separated, both methods may yield similar cluster assignments.

#Differences:
#Cluster shapes may appear different due to the transformation of axes in PCA.
#Number of clusters detected (if using algorithms like DBSCAN) may vary due to the change in density/variance distribution.
#Cluster centroids and boundaries may shift as PCA creates linear combinations of original features.


#3. Impact of Dimensionality Reduction on Clustering Performance
#Positive Impacts:
#Reduces computational complexity (especially for large datasets).
#Improves clustering stability by removing multicollinearity and noise.
#Enables clearer visualizations of clusters.

#Negative Impacts:
#PCA is a linear transformationâ€”non-linear patterns may be distorted or lost.
#Important features that donâ€™t contribute to top principal components might be ignored.
#Interpretation of clusters becomes harder since PCA components are abstract.
```

```python
#Task 6: Conclusion and Insights
#Using PCA before clustering is generally beneficial for visualization, noise reduction, and speed, especially in high-dimensional datasets. 
#However, for applications requiring direct interpretation of clusters (e.g., feature importance, business decisions), clustering on the original dataset
#might be more suitable. The best approach often involves testing both and comparing performance metrics (e.g., silhouette score, Davies-Bouldin index)
#and interpretability.


#Practical Implications of Using PCA and Clustering in Data Analysis
#Principal Component Analysis (PCA) and clustering are commonly used together in unsupervised learning tasks, especially for exploring and understanding
#large, complex datasets. Here are the practical implications of using them:
# 1.1. Improved Efficiency and Speed
 #2. Better Visualization and Interpretation
#3. Enhanced Clustering Quality (Sometimes)
#4. Loss of Interpretability
#5. Handling Multicollinearity
#6. Use in Preprocessing Pipelines


# Final Recommendation
#For most real-world datasets, especially when dealing with many features, 
#it's often best to use PCA for preprocessing before clustering. However, always evaluate the trade-off between interpretability
#and efficiency, and compare results with and without PCA using metrics like Silhouette Score or Davies-Bouldin Index.
```
